# What-if the temporal variability of dgemm durations was different?

In this experiment, we simulated several HPL runs with different matrix sizes. The simulated platform is the 32 Dahu nodes. In the model we use, we change the parameters for sigma ("residual" parameters) to have a defined [coefficient of variation](https://en.wikipedia.org/wiki/Coefficient_of_variation). For instance, to have a coefficient of $\lambda$, the value for `mnk_residual` is equal to the value for `mnk` times $\lambda$.

````python
import io
import zipfile
import os
import pandas
from plotnine import *
import plotnine
plotnine.options.figure_size = (12, 8)
import yaml
import warnings
warnings.simplefilter(action='ignore')

def get_yaml(archive_name, yaml_name):
    archive = zipfile.ZipFile(archive_name)
    return yaml.load(io.BytesIO(archive.read(yaml_name)))

def get_platform(archive_name):
    info = get_yaml(archive_name, 'info.yaml')
    expfiles = info['expfile']
    platform = [f for f in expfiles if f.endswith('xml')]
    assert len(platform) == 1
    return platform[0]

def read_csv(archive_name, file_name):
    archive = zipfile.ZipFile(archive_name)
    res = pandas.read_csv(io.BytesIO(archive.read(file_name)))
    res['filename'] = archive_name
    return res

def read_result(name):
    res = read_csv(name, 'results.csv')
    res['start_timestamp'] = pandas.to_datetime(res['start_timestamp'])
    res['start'] = res['start_timestamp'] - res['start_timestamp'].min()
    return res

def read_sim_result(name):
    archive = zipfile.ZipFile(name)
    result = pandas.read_csv(io.BytesIO(archive.read('results.csv')))
    result['platform'] = get_platform(name)
    result['filename'] = name
    info = get_yaml(name, 'info.yaml')
    expfiles = info['expfile']
    dgemm_file = [f for f in expfiles if f.endswith('.yaml')]
    assert len(dgemm_file) == 1
    dgemm_model = get_yaml(name, dgemm_file[0])
    comment = dgemm_model['metadata']['comment']
    assert comment.startswith('Modified the residuals to have a coefficient of variation of ')
    factor = float(comment[len('Modified the residuals to have a coefficient of variation of '):])
    result['variation_coefficient'] = factor
    return result

simulation_dir = 'dgemm_variability/3/'
simulation_files = [os.path.join(simulation_dir, f) for f in os.listdir(simulation_dir) if f.endswith('.zip')]

df = pandas.concat([read_sim_result(f) for f in simulation_files])
df.head()
````

### Checking the parameters

````python
name_exceptions = {'application_time', 'simulation_time', 'usr_time', 'sys_time', 'time', 'gflops', 'residual', 'cpu_utilization',
                   'dgemm_coefficient', 'dgemm_intercept', 'dtrsm_coefficient', 'dtrsm_intercept',
                   'stochastic_cpu', 'polynomial_dgemm', 'stochastic_network', 'heterogeneous_dgemm', 'platform', 'model', 'filename',
                   'simulation', 'slow_nodes',
                   'major_page_fault', 'minor_page_fault', 'matrix_size', 'mode',
                   'start_timestamp', 'stop_timestamp'}
colnames = set(df) - name_exceptions
df[list(colnames)].drop_duplicates()
````

````python
from IPython.display import display, Markdown

platforms = [(get_platform(f), zipfile.ZipFile(f).read(get_platform(f)).decode('ascii')) for f in simulation_files]
platforms = list(set(platforms))
assert(len(platforms)) == 1
for name, plat in platforms:
    display(Markdown('### %s' % name))
    display(Markdown('```xml\n%s\n```' % plat))
````

### Checking the patch in the simulation

````python
patches = set()
for row in df.iterrows():
    filename = row[1].filename
    repos = get_yaml(filename, 'info.yaml')['git_repositories']
    hpl = [repo for repo in repos if repo['path'] == 'hpl-2.2']
    assert len(hpl) == 1
    patches.add(hpl[0]['patch'])
assert len(patches) == 1
display(Markdown('```diff\n%s\n```' % patches.pop()))
````

## Impact on the predicted HPL performance

````python
tmp = df[df.matrix_size == 500000].copy()
max_perf = tmp[tmp.variation_coefficient == 0].gflops.max()
tmp['diff'] = (max_perf - tmp['gflops']) / max_perf * 100
tmp['var'] = tmp.variation_coefficient * 100
tmp['diff'] = tmp.apply(lambda row: '%4.1f%%' % row['diff'], axis=1)
tmp['var'] = tmp.apply(lambda row: '%2.f%%' % row['var'], axis=1)
tmp['text'] = 'CV=' + tmp['var'] + ' → drop=' + tmp['diff']
tmp['y'] = tmp['gflops']
tmp.loc[tmp.variation_coefficient == 0, 'y'] *= 1.02
````

````python
tmp['x'] = tmp['matrix_size'] * 1.14
plot = ggplot(df) + aes(x='matrix_size', y='gflops', color='factor(variation_coefficient)') + geom_point() + geom_line()
plot = plot + xlab('Matrix size') + ylab('Performance (Gflop/s)') + labs(color='Coefficient of variation')
plot = plot + theme_bw() + expand_limits(y=0) + expand_limits(x=600000)
plot = plot + geom_text(tmp, aes(x='x', y='y', label='text'))
margin = 0.02
plot = plot + geom_rect(xmin=tmp.matrix_size.min()*(1-margin), xmax=tmp.matrix_size.max()*(1+margin),
                        ymin=tmp.gflops.min()*(1-margin), ymax=tmp.gflops.max()*(1+margin),
                        fill='none', color='black', linetype='dotted', size=0.01)
plot
````

````python
plot = ggplot(df) + aes(x='variation_coefficient', y='time', color='factor(matrix_size)') + geom_point()
plot = plot + xlab('Coefficient of variation') + ylab('Total time (s)') + labs(color='Matrix size')
plot = plot + theme_bw() + expand_limits(y=0) + geom_smooth(df[df.variation_coefficient >= 0.02], method='lm')
plot
````

````python
base_time = df[df['variation_coefficient'] == 0].set_index('matrix_size')[['time']]
df = df.set_index('matrix_size').join(base_time, rsuffix='_base').reset_index()
df['additional_time'] = df['time'] - df['time_base']
````

````python
plot = ggplot(df) + aes(x='variation_coefficient', y='additional_time', color='factor(matrix_size)') + geom_point()
plot = plot + xlab('Coefficient of variation') + ylab('Additional time (s)') + labs(color='Matrix size')
plot = plot + theme_bw() + expand_limits(y=0) + geom_smooth(df[df.variation_coefficient >= 0.02], method='lm')
plot
````

## Impact on the simulation time

````python
df['simgrid_time'] = df['simulation_time'] - df['application_time']

plot = ggplot(df) + aes(x='matrix_size', color='factor(variation_coefficient)') + geom_point() + geom_line()
plot = plot + xlab('Matrix size') + ylab('Simulation time (s)') + labs(color='Coefficient of variation')
plot = plot + theme_bw() + expand_limits(y=0)

for y in ['Simulation', 'Application', 'Simgrid']:
    print(plot + aes(y='%s_time' % y.lower()) + ylab('%s time (s)' % y))
````
